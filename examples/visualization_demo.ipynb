# LPM Validation Visualization Demo

This notebook demonstrates how to use the `lpm_validation.visualization` module to analyze and visualize validation data.

## Data File Structure

CSV files are organized by simulator and car. Each file contains all variants of a specific car simulated using a specific simulator:

- **File naming**: `{Simulator}_{Car_Name}.csv`
  - Example: `JakubNet_BMW_IX.csv` contains all BMW_IX variants simulated using JakubNet
  - Example: `DES_Mitsubishi_Eclipse.csv` contains all Mitsubishi_Eclipse variants simulated using DES

- **Data location**: `examples/data/` directory contains example datasets

## Setup

First, import the necessary modules:

```python
from lpm_validation.visualization import (
    load_csv,
    load_dataset,
    load_multiple_datasets,
    CoefficientScatterPlot,
    ColorScheme
)

import pandas as pd
```

## Example 1: Load Specific CSV File

Load a single CSV file containing all variants of a specific car from a specific simulator.

```python
# Load BMW IX data simulated with JakubNet
# File: JakubNet_BMW_IX.csv contains all BMW_IX variants from JakubNet simulator
df = load_csv('data/JakubNet_BMW_IX.csv')
print(f"Loaded {len(df)} records")
print(f"Columns: {list(df.columns)}")
df.head()
```

```python
# Create scatter plot of CD values
plot = CoefficientScatterPlot(
    df,
    y_metric='Cd',
    color_by='Morph_Type',
    title='Drag Coefficient - BMW IX (JakubNet)'
)
plot.show()
```

## Example 2: Compare Same Car Across Different Simulators

Load data for the same car from different simulators and compare them on the same plot.

```python
# Load Mitsubishi Eclipse data from two different simulators
# File: JakubNet_Mitsubishi_Eclipse.csv - JakubNet simulator results
# File: DES_Mitsubishi_Eclipse.csv - DES simulator results
df_jakub = load_csv('data/JakubNet_Mitsubishi_Eclipse.csv')
df_des = load_csv('data/DES_Mitsubishi_Eclipse.csv')

print(f"JakubNet: {len(df_jakub)} records")
print(f"DES: {len(df_des)} records")
```

```python
# Create comparison plot
plot = CoefficientScatterPlot(
    [df_jakub, df_des],
    labels=['JakubNet', 'DES'],
    y_metric='Cd',
    color_by='Morph_Type',
    symbol_by='dataset',
    title='CD Comparison: JakubNet vs DES - Mitsubishi Eclipse'
)
plot.show()
```

## Example 4: Load Multiple Files from Directory

Load all CSV files from a directory (e.g., when you have multiple simulator/car combinations).

```python
# Load all CSV files from data directory
df_all = load_dataset('data')
print(f"Loaded {len(df_all)} total records from all files")
print(f"Unique cars: {df_all['Baseline_ID'].unique()}")
print(f"Unique simulators: {df_all['Simulator'].unique()}")
```

```python
# Plot all data colored by car, symbol by simulator
plot = CoefficientScatterPlot(
    df_all,
    y_metric='Cd',
    color_by='Baseline_ID',
    symbol_by='Simulator',
    title='Drag Coefficient - All Data'
)
plot.show()
```

## Example 5: Plot Against Morph Value

Visualize how coefficients change with morph parameter values.

```python
# Load Mitsubishi Eclipse data
df_morph = load_csv('data/JakubNet_Mitsubishi_Eclipse.csv')

# Filter to only one morph type for clarity
morph_type = 'Front Overhang'
df_filtered = df_morph[df_morph['Morph_Type'] == morph_type]

print(f"Analyzing morph type: {morph_type}")
print(f"Records: {len(df_filtered)}")
```

```python
# Plot CD vs Morph Value
plot = CoefficientScatterPlot(
    df_filtered,
    y_metric='Cd',
    x_axis='Morph_Value',
    show_baseline=False,
    title=f'CD vs {morph_type} - Mitsubishi Eclipse'
)
plot.show()
```

## Example 6: Compare Different Cars (Same Simulator)

Compare aerodynamic performance across different car models.

```python
# Load data for multiple cars from same simulator
df_bmw = load_csv('data/JakubNet_BMW_IX.csv')
df_mitsu = load_csv('data/JakubNet_Mitsubishi_Eclipse.csv')

print(f"BMW IX: {len(df_bmw)} records")
print(f"Mitsubishi Eclipse: {len(df_mitsu)} records")

# Create comparison plot
plot = CoefficientScatterPlot(
    [df_bmw, df_mitsu],
    labels=['BMW IX', 'Mitsubishi Eclipse'],
    y_metric='Cd',
    color_by='Morph_Type',
    symbol_by='dataset',
    title='CD Comparison: BMW IX vs Mitsubishi Eclipse (JakubNet)'
)
plot.show()
```

## Example 7: Save Plot to File

Export plots to HTML or static images.

```python
# Create a plot
df_all = load_dataset('data')
plot = CoefficientScatterPlot(
    df_all,
    y_metric='Cd',
    color_by='Baseline_ID',
    symbol_by='Simulator',
    title='Drag Coefficient - All Results'
)

# Save as HTML (interactive)
plot.save('cd_scatter_all.html')
print("Saved to cd_scatter_all.html")

# Save as PNG (requires kaleido)
try:
    plot.save('cd_scatter_all.png', format='png', width=1200, height=600)
    print("Saved to cd_scatter_all.png")
except Exception as e:
    print(f"Could not save PNG (install kaleido): {e}")
```
